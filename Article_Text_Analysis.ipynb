```python
# Load the Excel file with URLs
excel_file_path = r"C:\Users\Harshit\Downloads\Input.xlsx"
df = pd.read_excel(excel_file_path)
url_list = df['URL'].tolist()

# Create a new Excel workbook for the output
output_workbook = Workbook()
output_sheet = output_workbook.active
output_sheet.append([
    "URL", "Article Title", "Article Text",
    "Positive Score", "Negative Score", "Polarity Score", "Subjectivity Score",
    "Avg Sentence Length", "Percentage of Complex Words", "FOG Index",
    "Avg Number of Words per Sentence", "Complex Word Count", "Word Count",
    "Syllable per Word", "Personal Pronouns", "Avg Word Length"
])

# Iterate through each URL and scrape data
for url in url_list:
    response = requests.get(url)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the article title using different class names
    possible_title_class_names = ['entry-title', 'tdb-title-text']
    article_title = None
    for class_name in possible_title_class_names:
        article_title_element = soup.find('h1', class_=class_name)
        if article_title_element:
            article_title = article_title_element.text
            break

    if article_title is None:
        print(f"Article title not found for {url}")
        continue

    # Find the article text based on the provided class name
    article_text = ''
    article_text_elements = soup.find_all('div', class_='td-pb-span8 td-main-content')
    for element in article_text_elements:
        article_text += element.get_text()

    # Perform textual analysis using TextBlob
    blob = TextBlob(article_text)
    
    # Sentiment Analysis
    sia = SentimentIntensityAnalyzer()
    sentiment_scores = sia.polarity_scores(article_text)
    positive_score = sentiment_scores['pos']
    negative_score = sentiment_scores['neg']
    polarity_score = sentiment_scores['compound']
    subjectivity_score = blob.sentiment.subjectivity
    
    # Calculate average sentence length
    sentences = blob.sentences
    if len(sentences) == 0:
        avg_sentence_length = 0
    else:
        avg_sentence_length = sum(len(sentence.words) for sentence in sentences) / len(sentences)
    
    # Calculate complex word count based on syllables
    word_list = blob.words
    complex_word_count = sum(1 for word in word_list if syllables.estimate(word) >= 3)
    
    # Calculate FOG Index
    if len(word_list) == 0:
        fog_index = 0
    else:
        fog_index = 0.4 * (avg_sentence_length + (complex_word_count / len(word_list)) * 100)
    
    # Calculate percentage of complex words
    if len(word_list) == 0:
        percentage_complex_words = 0
    else:
        percentage_complex_words = (complex_word_count / len(word_list)) * 100
    
    # Compute average number of words per sentence
    avg_words_per_sentence = len(word_list) / len(sentences) if len(sentences) > 0 else 0
    
    # Compute syllables per word
    syllables_per_word = syllables.estimate(article_text) / len(word_list) if len(word_list) > 0 else 0
    
    # Count personal pronouns
    personal_pronouns = len([word for word in blob.words if word.lower() in ['i', 'me', 'my', 'mine', 'we', 'us', 'our', 'ours']])
    
    # Compute average word length
    words = re.findall(r'\b\w+\b', article_text)
    avg_word_length = sum(len(word) for word in words) / len(words) if len(words) > 0 else 0

    # Add data to the output sheet
    output_sheet.append([
        url, article_title, article_text,
        positive_score, negative_score, polarity_score, subjectivity_score,
        avg_sentence_length, percentage_complex_words, fog_index,
        avg_words_per_sentence, complex_word_count, len(word_list),
        syllables_per_word, personal_pronouns, avg_word_length
    ])

# Save the output Excel file
output_excel_path = 'output_with_metrics.xlsx'
output_workbook.save(output_excel_path)

print("Textual analysis results with metrics saved to", output_excel_path)
```
