# Getting Data Directly From Link

```python
url="https://www.amazon.in/Noise-Wireless-Earbuds-Playtime-Technology/dp/B0979DG42T?ref_=Oct_DLandingS_D_47841916_88"
c = requests.get(url)
htmlContent = c.content
#print(htmlContent)
```

# Using Beautiful Soup Libarary

```python
soup=BeautifulSoup(htmlContent,"html.parser")
```

```python
#print(soup.prettify())
print(type(soup))
```

# HTML Tree Travel

```python
#Get the title of the web Page
title=soup.title
print(title)
```

```python
print(type(title))
```

```python
print(type(title.string))
```

```python
#Getting Paragraphs From The Page
paras=soup.find_all("p")
#print(paras)
```

```python
#Getting Anchor tags From The Page
anchors=soup.find_all("a")
#print(anchors)
```

```python
print(soup.find("p"))# Get first Paragraph from Page
```

```python
print(soup.find("p")["class"])# Get classes of any para
```

```python
# Get all the elemets with the class lead
print(soup.find_all("p",class_="lead"))
```

```python
print(soup.find("p").get_text())
```
